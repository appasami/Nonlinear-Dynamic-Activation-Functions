```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.optimizers import AdamW
from sklearn.utils import shuffle
import matplotlib.pyplot as plt
import numpy as np
import os
from sklearn.metrics import classification_report
from tabulate import tabulate

# Global variables
SAVE = False
SEED = 111

# Setting seed for consistent results
tf.keras.utils.set_random_seed(SEED)
tf.random.set_seed(SEED)
np.random.seed(SEED)

# Data Visualization updates
plt.rcParams["figure.figsize"] = (16, 10)
plt.rcParams.update({'font.size': 14})

# Data Classifications
CLASS_TYPES = ['pituitary', 'notumor', 'meningioma', 'glioma']
N_TYPES = len(CLASS_TYPES)

# Image size
image_size = (150, 150)

# Training batch size
batch_size = 16

# Training epochs (reduced for quicker demonstration of multiple activations)
epochs = 64 # Reduced for demonstration purposes

# Image shape: height, width, RBG
image_shape = (image_size[0], image_size[1], 3)

# Function for importing data
def get_data_labels(directory, do_shuffle=True, random_state=0):
    """
    Function used for going into the main training directory
    whose directory has sub-class-types.
    """
    from sklearn.utils import shuffle
    import os

    # Lists to store data and labels
    data_path = []
    data_labels = []

    for label in os.listdir(directory):
        label_dir = os.path.join(directory, label)

        # Avoid MacOS storing path
        if not os.path.isdir(label_dir):
            continue

        # Going into each folder and getting image path
        for image in os.listdir(label_dir):
            image_path = os.path.join(label_dir, image)
            data_path.append(image_path)
            data_labels.append(label)

    if do_shuffle:
        data_path, data_labels = shuffle(data_path, data_labels, random_state=random_state)

    return data_path, data_labels

# Setting up file paths for training and testing
USER_PATH = "/kaggle/input/brain-tumor-mri-dataset"
train_dir = USER_PATH + "/Training/"
test_dir = USER_PATH + "/Testing/"
save_path = './Optimizer_MRI_AdamW' # This path is not used in the current code, but kept from original

# Getting data using above function
train_paths, train_labels = get_data_labels(train_dir)
test_paths, test_labels = get_data_labels(test_dir)

# Data augmentation and preprocessing
train_datagen = ImageDataGenerator(rescale=1./255,
                                   rotation_range=10,
                                   brightness_range=(0.9, 1.1),
                                   width_shift_range=0.002,
                                   height_shift_range=0.002,
                                   shear_range=10,
                                   zoom_range=0,
                                   horizontal_flip=True,
                                   vertical_flip=False,
                                   fill_mode="nearest")

train_generator = train_datagen.flow_from_directory(train_dir,
                                                    target_size=image_size,
                                                    batch_size=batch_size,
                                                    class_mode="categorical",
                                                    seed=SEED)

test_datagen = ImageDataGenerator(rescale=1./255)

test_generator = test_datagen.flow_from_directory(test_dir,
                                                   target_size=image_size,
                                                   batch_size=batch_size,
                                                   class_mode="categorical",
                                                   shuffle=False,
                                                   seed=SEED)

# Calculate steps per epoch
steps_per_epoch = train_generator.samples // train_generator.batch_size
validation_steps = test_generator.samples // test_generator.batch_size

# --- Output Images and Labels Visualization ---
def plot_sample_predictions(model, test_generator, categories, num_samples=9, figsize=(12, 8)):
    """
    Nice display of prediction samples to see CNN predictions
    for classification.
    """
    predictions = model.predict(test_generator)
    predicted_categories = np.argmax(predictions, axis=1)
    true_categories = test_generator.classes

    test_images = np.array(test_generator.filepaths)
    sample_indices = np.random.choice(len(test_images), size=num_samples, replace=False)
    sample_images = test_images[sample_indices]
    sample_predictions = [categories[predicted_categories[i]] for i in sample_indices]
    sample_true_labels = [categories[true_categories[i]] for i in sample_indices]

    plt.figure(figsize=figsize)
    for i, image_path in enumerate(sample_images):
        plt.subplot(3, 3, i + 1)
        img = plt.imread(image_path)
        plt.imshow(img)
        plt.axis("off")

        prediction_color = 'green' if sample_predictions[i] == sample_true_labels[i] else 'red'
        plt.title(f"Pred: {sample_predictions[i]}\nTrue: {sample_true_labels[i]}", color=prediction_color, fontsize=10)

    plt.tight_layout()
    plt.show()

# --- Custom Activation Functions ---
class CustomActivations:
    @staticmethod
    def leaky_relu(x):
        return tf.keras.layers.LeakyReLU(alpha=0.01)(x)

    @staticmethod
    def prelu(x):
        return tf.keras.layers.PReLU()(x)

    @staticmethod
    def swish(x):
        return tf.keras.activations.swish(x)

    @staticmethod
    def mish(x):
        return x * tf.math.tanh(tf.math.softplus(x))

    @staticmethod
    def hard_swish(x):
        return x * tf.keras.activations.relu(x + 3, max_value=6) / 6

    @staticmethod
    def hard_sigmoid(x):
        return tf.keras.activations.relu(x + 2.5, max_value=5.0) / 5.0

    @staticmethod
    def thresholded_relu(x):
        return tf.keras.layers.ThresholdedReLU(theta=1.0)(x)

    @staticmethod
    def bent_identity(x):
        return (tf.sqrt(x**2 + 1) - 1) / 2 + x

    @staticmethod
    def arctan(x):
        return tf.math.atan(x)

    @staticmethod
    def sinusoid(x):
        return tf.math.sin(x)

    @staticmethod
    def sinc(x):
        # Handle division by zero for x=0
        return tf.where(tf.equal(x, 0), tf.constant(1.0, dtype=x.dtype), tf.math.sin(x) / x)

    @staticmethod
    def cube(x):
        return x**3

    # New custom activations
    @staticmethod
    def lisht(x):
        return x * tf.math.tanh(x)

    @staticmethod
    def snake(x, a=0.5):
        return x + a * tf.sin(x)**2

    @staticmethod
    def gaussian(x):
        return tf.exp(-x**2)

    @staticmethod
    def tanhshrink(x):
        return x - tf.math.tanh(x)

    @staticmethod
    def softshrink(x, lmbd=0.5):
        return tf.where(x > lmbd, x - lmbd, tf.where(x < -lmbd, x + lmbd, 0.0))

    @staticmethod
    def hardshrink(x, lmbd=0.5):
        return tf.where(x > lmbd, x, tf.where(x < -lmbd, x, 0.0))

    @staticmethod
    def celu(x, alpha=1.0):
        return tf.where(x > 0, x, alpha * (tf.exp(x / alpha) - 1))

    @staticmethod
    def mexican_hat(x):
        return (1 - x**2) * tf.exp(-0.5 * x**2)

    @staticmethod
    def morlet(x, s=1.0):
        return tf.cos(1.75 * x / s) * tf.exp(-0.5 * (x / s)**2)

    @staticmethod
    def ftsplash(x):
        return x * tf.tanh(tf.math.log(1 + tf.exp(x)))

    @staticmethod
    def erfaf(x):
        return tf.math.erf(x) # Error Function Activation Function

    @staticmethod
    def logish(x):
        return tf.math.log(tf.math.cosh(x))

    @staticmethod
    def squareplus(x):
        return 0.5 * (x + tf.sqrt(x**2 + 4))

    @staticmethod
    def logarithmic(x, base=tf.exp(1.0)):
        # Added a small epsilon to x for numerical stability to prevent log(0)
        return tf.where(x > 1e-7, tf.math.log(x) / tf.math.log(base), tf.constant(-float('inf'), dtype=x.dtype))

    @staticmethod
    def exponential(x):
        return tf.exp(x)

    @staticmethod
    def bipolar_sigmoid(x):
        return (1 - tf.exp(-x)) / (1 + tf.exp(-x))

    @staticmethod
    def complementary_log_log(x):
        # Inverse of Weibull distribution's CDF
        return 1 - tf.exp(-tf.exp(x))

    @staticmethod
    def phish(x):
        return x * tf.math.tanh(tf.math.softplus(x)) # Similar to Mish, often refers to Mish

    # RReLU as a Keras Layer due to its training-dependent behavior
    class RReLU(tf.keras.layers.Layer):
        """
        Randomized Leaky ReLU activation function.
        The negative slope 'alpha' is randomly chosen from a uniform distribution
        during training and fixed to its average during inference.
        """
        def __init__(self, lower_bound=0.125, upper_bound=0.333, **kwargs):
            super(CustomActivations.RReLU, self).__init__(**kwargs)
            self.lower_bound = lower_bound
            self.upper_bound = upper_bound
            self.alpha_inference = (lower_bound + upper_bound) / 2.0 # Average for inference

        def call(self, inputs, training=None):
            if training:
                # Random alpha for each training step
                alpha = tf.random.uniform(shape=tf.shape(inputs),
                                          minval=self.lower_bound,
                                          maxval=self.upper_bound,
                                          dtype=inputs.dtype)
                return tf.where(inputs > 0, inputs, alpha * inputs)
            else:
                # Fixed alpha for inference
                return tf.where(inputs > 0, inputs, self.alpha_inference * inputs)

        def get_config(self):
            config = super(CustomActivations.RReLU, self).get_config()
            config.update({
                "lower_bound": self.lower_bound,
                "upper_bound": self.upper_bound,
            })
            return config

    # --- NEW CUSTOM ACTIVATION FUNCTIONS ---
    @staticmethod
    def slrelu(x, alpha=0.01):
        """
        Smoothed Leaky ReLU (SLReLU) activation function.
        Uses a sigmoid to smoothly transition the negative slope.
        """
        # Ensure alpha is a Tensor for broadcasting if x is multi-dimensional
        alpha = tf.cast(alpha, dtype=x.dtype)
        sigmoid_val = tf.math.sigmoid(x * 10.0) # Multiply by a factor for sharper (but still smooth) transition
        return x * sigmoid_val + alpha * x * (1 - sigmoid_val)

    class AGELU(tf.keras.layers.Layer):
        """
        Adaptive GELU (AGELU) activation function with a learnable beta parameter.
        Inspired by GELU with an adaptive scaling.
        """
        def __init__(self, initial_beta=1.0, **kwargs):
            super().__init__(**kwargs)
            self.initial_beta = initial_beta

        def build(self, input_shape):
            self.beta = self.add_weight(name='beta',
                                        shape=(), # Scalar
                                        initializer=tf.keras.initializers.Constant(self.initial_beta),
                                        trainable=True,
                                        dtype=tf.float32)
            super().build(input_shape)

        def call(self, inputs):
            # Using the tanh approximation for GELU
            return 0.5 * inputs * (1.0 + tf.math.tanh(self.beta * (tf.sqrt(2.0 / np.pi) * (inputs + 0.044715 * tf.pow(inputs, 3)))))

        def get_config(self):
            config = super().get_config()
            config.update({
                "initial_beta": self.initial_beta,
            })
            return config

    @staticmethod
    def qslu(x, alpha=0.01):
        """
        Quadratic Sigmoid Linear Unit (QSLU) activation function.
        Blends a linear term with a sigmoid-scaled quadratic influence.
        """
        alpha = tf.cast(alpha, dtype=x.dtype)
        return x + alpha * x**2 * tf.math.sigmoid(x)


# --- Function to build, compile, train, and plot for a given activation ---
def run_model_with_activation(activation_func_name, activation_param):
    print(f"\n--- Running Model with {activation_func_name} Activation ---")

    model = models.Sequential()

    # Counter for unique layer names within each model build
    # Reset for each new model
    layer_counter = {}

    def get_unique_name(prefix):
        if prefix not in layer_counter:
            layer_counter[prefix] = 0
        name = f"{prefix}_{layer_counter[prefix]}"
        layer_counter[prefix] += 1
        return name

    # Helper to add Conv2D layer with correct activation handling
    def add_conv_layer(filters, kernel_size, input_shape=None, layer_idx=0):
        if isinstance(activation_param, str): # Activation is a string (e.g., 'relu', 'elu')
            model.add(layers.Conv2D(filters, kernel_size, activation=activation_param, input_shape=input_shape, name=get_unique_name(f"conv_{activation_func_name.lower().replace(' ', '_')}_{layer_idx}")))
        elif isinstance(activation_param, type) and issubclass(activation_param, tf.keras.layers.Layer): # Activation is a Keras Layer class (e.g., CustomActivations.RReLU, CustomActivations.AGELU)
            model.add(layers.Conv2D(filters, kernel_size, input_shape=input_shape, name=get_unique_name(f"conv_pre_{activation_func_name.lower().replace(' ', '_')}_{layer_idx}")))
            # For Keras Layer classes like RReLU, AGELU, instantiate it.
            if activation_func_name.lower() == 'rrelu':
                new_layer = activation_param(lower_bound=0.1, upper_bound=0.3, name=get_unique_name(f"{activation_func_name.lower().replace(' ', '_')}_{layer_idx}"))
            elif activation_func_name.lower() == 'agelu':
                new_layer = activation_param(initial_beta=1.0, name=get_unique_name(f"{activation_func_name.lower().replace(' ', '_')}_{layer_idx}"))
            else: # For other Keras Layer classes like LeakyReLU, PReLU, ThresholdedReLU without specific custom params
                new_layer = activation_param(name=get_unique_name(f"{activation_func_name.lower().replace(' ', '_')}_{layer_idx}"))
            model.add(new_layer)
        elif callable(activation_param): # Activation is a custom callable function (e.g., CustomActivations.mish, CustomActivations.sinc)
            model.add(layers.Conv2D(filters, kernel_size, input_shape=input_shape, name=get_unique_name(f"conv_pre_{activation_func_name.lower().replace(' ', '_')}_{layer_idx}")))
            # Use Activation layer with a unique name
            model.add(layers.Activation(activation_param, name=get_unique_name(f"{activation_func_name.lower().replace(' ', '_')}_activation_{layer_idx}")))
        else: # Fallback or error if unrecognized
            raise ValueError(f"Unsupported activation parameter type for Conv2D: {type(activation_param)}")

    # Helper to add Dense layer with correct activation handling
    def add_dense_layer(units, layer_idx=0):
        if isinstance(activation_param, str):
            model.add(layers.Dense(units, activation=activation_param, name=get_unique_name(f"dense_{activation_func_name.lower().replace(' ', '_')}_{layer_idx}")))
        elif isinstance(activation_param, type) and issubclass(activation_param, tf.keras.layers.Layer):
            model.add(layers.Dense(units, name=get_unique_name(f"dense_pre_{activation_func_name.lower().replace(' ', '_')}_{layer_idx}")))
            if activation_func_name.lower() == 'rrelu':
                new_layer = activation_param(lower_bound=0.1, upper_bound=0.3, name=get_unique_name(f"{activation_func_name.lower().replace(' ', '_')}_dense_{layer_idx}"))
            elif activation_func_name.lower() == 'agelu':
                new_layer = activation_param(initial_beta=1.0, name=get_unique_name(f"{activation_func_name.lower().replace(' ', '_')}_dense_{layer_idx}"))
            else:
                new_layer = activation_param(name=get_unique_name(f"{activation_func_name.lower().replace(' ', '_')}_dense_{layer_idx}"))
            model.add(new_layer)
        elif callable(activation_param):
            model.add(layers.Dense(units, name=get_unique_name(f"dense_pre_{activation_func_name.lower().replace(' ', '_')}_{layer_idx}")))
            # Use Activation layer with a unique name
            model.add(layers.Activation(activation_param, name=get_unique_name(f"{activation_func_name.lower().replace(' ', '_')}_dense_activation_{layer_idx}")))
        else:
            raise ValueError(f"Unsupported activation parameter type for Dense: {type(activation_param)}")

    # Convolutional layer 1
    add_conv_layer(32, (4, 4), input_shape=image_shape, layer_idx=1)
    model.add(layers.MaxPooling2D(pool_size=(4, 4), name=get_unique_name(f"max_pooling_{activation_func_name.lower().replace(' ', '_')}_1")))

    # Convolutional layer 2
    add_conv_layer(64, (4, 4), layer_idx=2)
    model.add(layers.MaxPooling2D(pool_size=(4, 4), name=get_unique_name(f"max_pooling_{activation_func_name.lower().replace(' ', '_')}_2")))

    # Convolutional layer 3
    add_conv_layer(128, (4, 4), layer_idx=3)
    model.add(layers.Flatten(name=get_unique_name(f"flatten_{activation_func_name.lower().replace(' ', '_')}")))

    # Full connect layers
    add_dense_layer(512, layer_idx=1)
    model.add(layers.Dropout(0.35, seed=SEED, name=get_unique_name(f"dropout_{activation_func_name.lower().replace(' ', '_')}")))
    model.add(layers.Dense(N_TYPES, activation="softmax", name=get_unique_name(f"output_{activation_func_name.lower().replace(' ', '_')}"))) # Output layer always softmax for classification

    model.summary()

    # Display the kernel size for each convolutional layer
    for layer in model.layers:
        if isinstance(layer, layers.Conv2D):
            print(f"Layer {layer.name} has kernel size: {layer.kernel_size}")

    optimizer = AdamW(learning_rate=0.001, beta_1=0.869, beta_2=0.995)
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics= ['accuracy'])

    model_es = EarlyStopping(monitor='loss', min_delta=1e-9, patience=8, verbose=True)
    model_rlr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, verbose=True)

    history = model.fit(train_generator,
                        steps_per_epoch=steps_per_epoch,
                        epochs=epochs,
                        validation_data=test_generator,
                        validation_steps=validation_steps,
                        callbacks=[model_es, model_rlr],
                        verbose=0) # Set verbose to 0 to avoid massive output

    # Get predictions
    y_true = test_generator.classes
    y_pred_prob = model.predict(test_generator, steps=int(np.ceil(test_generator.samples / test_generator.batch_size)), verbose=0)
    y_pred = np.argmax(y_pred_prob, axis=1) # Get predicted class labels

    # --- Print Classification Report ---
    print(f"\n--- Classification Report for {activation_func_name} Activation ---")
    report = classification_report(y_true, y_pred, digits=6, target_names=CLASS_TYPES, output_dict=True)
    print(classification_report(y_true, y_pred, digits=6, target_names=CLASS_TYPES))

    # Plot the training and validation accuracy and loss
    _, ax = plt.subplots(ncols=2, figsize=(15, 6))

    ax[0].plot(history.history['accuracy'], linewidth=2)
    ax[0].plot(history.history['val_accuracy'], linewidth=2)
    ax[0].set_title(f'Model Accuracy ({activation_func_name})')
    ax[0].set_xlabel('Epoch')
    ax[0].set_ylabel('Accuracy')
    ax[0].set_ylim(0, 1)
    ax[0].legend(['Train', 'Validation'])
    ax[0].grid(alpha=0.2)

    ax[1].plot(history.history['loss'], linewidth=2)
    ax[1].plot(history.history['val_loss'], linewidth=2)
    ax[1].set_title(f'Model Loss ({activation_func_name})')
    ax[1].set_xlabel('Epoch')
    ax[1].set_ylabel('Loss')
    ax[1].set_ylim(0, 1)
    ax[1].legend(['Train', 'Validation'])
    ax[1].grid(alpha=0.2)
    plt.show()

    # Plot sample predictions
    plot_sample_predictions(model, test_generator, CLASS_TYPES)

    # Return the metrics
    return {
        'Activation': activation_func_name,
        'Accuracy': report['accuracy'],
        'Weighted Avg Precision': report['weighted avg']['precision'],
        'Weighted Avg Recall': report['weighted avg']['recall'],
        'Weighted Avg F1-score': report['weighted avg']['f1-score'],
        'Macro Avg Precision': report['macro avg']['precision'],
        'Macro Avg Recall': report['macro avg']['recall'],
        'Macro Avg F1-score': report['macro avg']['f1-score'],
    }

# --- Activation Functions Dictionary (Now sorted alphabetically by key) ---
ACTIVATION_FUNCTIONS = {
###    'arctan': CustomActivations.arctan,
###    'bent_identity': CustomActivations.bent_identity,
###    'bipolar_sigmoid': CustomActivations.bipolar_sigmoid,
###    'celu': CustomActivations.celu,
###    # 'complementary_log_log': CustomActivations.complementary_log_log,
###    # 'cube': CustomActivations.cube,
###    'elu': 'elu',
###    'erfaf': CustomActivations.erfaf,
###    # 'exponential': CustomActivations.exponential,
###    'ftsplash': CustomActivations.ftsplash,
###    # 'gaussian': CustomActivations.gaussian,
###    'gelu': 'gelu',
###    # 'hard_sigmoid': CustomActivations.hard_sigmoid,
###    'hard_swish': CustomActivations.hard_swish,
###    # 'hardshrink': CustomActivations.hardshrink,
###    'leaky_relu': CustomActivations.leaky_relu,
###    'lisht': CustomActivations.lisht,
###    # 'logarithmic': CustomActivations.logarithmic,
###    'logish': CustomActivations.logish,
###    # 'mexican_hat': CustomActivations.mexican_hat,
###    'mish': CustomActivations.mish,
###    'morlet': CustomActivations.morlet,
###    'phish': CustomActivations.phish,
###    # 'prelu': CustomActivations.prelu, # PReLU is already a Keras Layer in CustomActivations
###    'relu': 'relu',
###    'rrelu': CustomActivations.RReLU, # RReLU is a Keras Layer class
###    'selu': 'selu',
###    'sigmoid': 'sigmoid',
###    # 'sinc': CustomActivations.sinc,
###    'sinusoid': CustomActivations.sinusoid,
###    'snake': CustomActivations.snake,
###    'softplus': 'softplus',
###    # 'softshrink': CustomActivations.softshrink,
###    'softsign': 'softsign',
###    # 'squareplus': CustomActivations.squareplus,
###    'swish': CustomActivations.swish,
###    'tanh': 'tanh',
    # 'tanhshrink': CustomActivations.tanhshrink,
    # 'thresholded_relu': CustomActivations.thresholded_relu,
    # --- ADDED NEW CUSTOM ACTIVATION FUNCTIONS ---
    'slrelu': CustomActivations.slrelu, # Smoothed Leaky ReLU
    'agelu': CustomActivations.AGELU,  # Adaptive GELU (Keras Layer)
    'qslu': CustomActivations.qslu,   # Quadratic Sigmoid Linear Unit
}

# --- List to store results from all models ---
all_model_metrics = []

# --- Run the comparison and collect metrics ---
# Iterate through the sorted items of the dictionary
for name, activation_param in sorted(ACTIVATION_FUNCTIONS.items()):
    metrics = run_model_with_activation(name, activation_param)
    all_model_metrics.append(metrics)

print("\n--- All activation function tests completed ---")

# --- Display Consolidated Metrics in a Table ---
if all_model_metrics:
    headers = [
        "Activation",
        "Accuracy",
        "Weighted Avg Precision",
        "Weighted Avg Recall",
        "Weighted Avg F1-score",
        "Macro Avg Precision",
        "Macro Avg Recall",
        "Macro Avg F1-score",
    ]
    table_data = []
    for m in all_model_metrics:
        table_data.append([
            m['Activation'],
            f"{m['Accuracy']:.4f}",
            f"{m['Weighted Avg Precision']:.4f}",
            f"{m['Weighted Avg Recall']:.4f}",
            f"{m['Weighted Avg F1-score']:.4f}",
            f"{m['Macro Avg Precision']:.4f}",
            f"{m['Macro Avg Recall']:.4f}",
            f"{m['Macro Avg F1-score']:.4f}",
        ])
    print("\n--- Consolidated Model Performance Metrics ---")
    print(tabulate(table_data, headers=headers, tablefmt="grid"))
else:
    print("\nNo model metrics were collected.")
```
